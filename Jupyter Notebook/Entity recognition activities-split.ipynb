```python
import pandas as pd
import spacy
import csv
```

#### import spacy english model

```python
nlp = spacy.load('en_core_web_sm')
```

#### add activity reviews using pandas dataframe

```python
activity = pd.read_csv(r'C:\Users\Dell\Documents\MASTER CDE\RISE INTERNSHIP\Rochamukti Rizcanofana\Data\TripAdvisorDataNov1st2018\dataNov1st\reviews_activities_cleaned.csv')
activity.head()
```

#### Hagia sophia- Turkey

```python
#select single activity and convert to string
hagia = activity[activity['r_property'] == 'Hagia Sophia Museum / Church']
s_hagia = '.'.join(hagia['r_text'].tolist())
doc = nlp(s_hagia)
#display entities using Displacy
from spacy import displacy
displacy.render(doc, style='ent', jupyter = True)
```

```python
def hagia_save_csv():
    #creating new column for storing the review
    hagia_df = pd.DataFrame(hagia['r_text'])
    hagia_df.columns = ['review']
    #Extract named entities, and beginning, middle and end characters
    def extract_named_ents(review):
        return [(ent.text, ent.label_) for ent in nlp(review).ents]
    #Create new column in data frame with named entity
    def add_named_ents(hagia_df):
        hagia_df['named_ents'] = hagia_df['review'].apply(extract_named_ents)
    add_named_ents(hagia_df)
    display(hagia_df)
    #export dataframe to csv
    hagia_df.to_csv(r'C:\Users\Dell\Desktop\hagiasophia2.csv',index = False, header=True)
hagia_save_csv()
```
